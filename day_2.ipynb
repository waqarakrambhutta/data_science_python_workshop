```python
# import libraries
import pandas as pd
import seaborn as sns
```

```python
# load data
df = sns.load_dataset("titanic")
```

```python
# glimpse data
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
# null values 
null_value = df.isnull()
null_value
```

```python
# checking percentage of null value
per_null = (null_value/df.shape[0])*100
per_null
```

```python
#drop deck column
df.drop("deck",axis=1,inplace=True)
```

```python
df.columns
```

```python
# removing mull values of embarked column
df.dropna(subset=['embarked','embark_town'], inplace=True)
```

```python
#filling age null with mean
df['age'].fillna(df['age'].mean(),inplace=True)
```

```python
# checking null values again
df.isnull().sum()
```

```python
# rmoving duplicate values
df.duplicated().sum()
```

```python
# EDA -- Descriptive Analysis
##1-- central tendency ( mean, median, mode, ...)
df['fare'].mean()
##2-- Dispersion(range,variance,standard deviation,percentile,....)
df['fare'].std()

df.describe()
```
